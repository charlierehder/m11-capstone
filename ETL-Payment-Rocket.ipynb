```
#read in file
df = spark.read.options(header='True').csv('/mnt/alysia/Paymentetl/payments.csv')
display(df)
```

```
# print column names
print(df.columns)
```

```
# dropping unnecessary columns as documented in Exercise-Initial-Capstone-Data-Sources
df = df.select('facility_id', 'state', 'payment_category', 'payment', 'lower_estimate', 'higher_estimate', 'payment_footnote', 'value_of_care_category', 'value_of_care_footnote')
df.show()
display(df)
```

```
from pyspark.sql.functions import *
# dropping rows with payment not available 
# filter rows where payment footnote and value of care footnote are null
df2 = df.filter(col("payment") != "Not Available")
df2.show(truncate=False)
display(df2)
```

```
# removing dollar sign from values, casting from string to double data type
import pyspark.sql.functions as f

df2 = df2.withColumn('payment', f.regexp_replace('payment', '[$,]', '').cast('double'))
df2 = df2.withColumn('higher_estimate', f.regexp_replace('higher_estimate', '[$,]', '').cast('double'))
df2 = df2.withColumn('lower_estimate', f.regexp_replace('lower_estimate', '[$,]', '').cast('double'))
display(df2)
```

```
# dropping unnecessary columns now that we have filtered out footnote entries
df = df2.drop('payment_footnote', 'value_of_care_footnote')
df.show()
display(df)
```

```
# print column names
print(df.columns)
```

```
# renaming columns for SQL database/readability
# confirming dataframe accepted column name changes

df = df.withColumnRenamed("facility_id","Facility_ID").withColumnRenamed("payment","Payment").withColumnRenamed("payment_category","Payment_Category_Str").withColumnRenamed("lower_estimate","Lower_Estimate").withColumnRenamed("higher_estimate","Higher_Estimate").withColumnRenamed("value_of_care_category","Value_Of_Care_Category_Str")
df.show()
display(df)
```

```
# view current data types
df.printSchema()
```

```
# map out payment category values to numeric
from pyspark.sql.functions import col, when

df = df.withColumn("Payment_Category", when(col("Payment_Category_Str")=='No Different Than the National Average Payment', 0).when(col("Payment_Category_Str")=='Greater Than the National Average Payment', 1).otherwise(-1))
df.show()
df = df.drop('Payment_Category_Str')
df.show()
display(df)
```

```
# map out value of care category values to numeric
from pyspark.sql.functions import col, when

df = df.withColumn("Value_Of_Care_Category", when(col("Value_Of_Care_Category_Str")=='Average Mortality and Average Payment', 0).when(col("Value_Of_Care_Category_Str")=='Average Mortality and Higher Payment', 0).when(col("Value_Of_Care_Category_Str")=='Average Mortality and Lower Payment', 0).when(col("Value_Of_Care_Category_Str")=='Average Complications and Average Payment', 0).when(col("Value_Of_Care_Category_Str")=='Average Complications and Higher Payment', 0).when(col("Value_Of_Care_Category_Str")=='Average Complications and Lower Payment', 0).when(col("Value_Of_Care_Category_Str")=='Better Mortality and Average Payment', 1).when(col("Value_Of_Care_Category_Str")=='Better Mortality and Higher Payment', 1).when(col("Value_Of_Care_Category_Str")=='Better Mortality and Lower Payment', 1).when(col("Value_Of_Care_Category_Str")=='Better Complications and Average Payment', 1).when(col("Value_Of_Care_Category_Str")=='Better Complications and Higher Payment', 1).when(col("Value_Of_Care_Category_Str")=='Better Complications and Lower Payment', 1).when(col("Value_Of_Care_Category_Str")=='Worse Mortality and Average Payment', -1).when(col("Value_Of_Care_Category_Str")=='Worse Mortality and Higher Payment', -1).when(col("Value_Of_Care_Category_Str")=='Worse Mortality and Lower Payment', -1).when(col("Value_Of_Care_Category_Str")=='Worse Complications and Average Payment', -1).when(col("Value_Of_Care_Category_Str")=='Worse Complications and Higher Payment', -1).when(col("Value_Of_Care_Category_Str")=='Worse Complications and Lower Payment', -1).otherwise(-1))
df.show()
df = df.drop('Value_Of_Care_Category_Str')
df.show()
display(df)
```

```
# double check category data types are int
df.printSchema()
```

```
#check, no missing values/nulls, no imputing needed
display(df)
```

```
# group by facility id and average out other columns for possible visualizations
facilityDf = df.groupBy("Facility_ID").avg("Payment", "Lower_Estimate", "Higher_Estimate", "Payment_Category", "Value_Of_Care_Category")
facilityDf.show()
# check grouped-by dataframe
display(facilityDf)
```

```
# renaming columns for SQL database/readability
# confirming dataframe accepted column name changes

facilityDf = facilityDf.withColumnRenamed("avg(Payment)","Payment").withColumnRenamed("avg(Lower_Estimate)","Lower_Estimate").withColumnRenamed("avg(Higher_Estimate)","Higher_Estimate").withColumnRenamed("avg(Payment_Category)","Payment_Category").withColumnRenamed("avg(Value_Of_Care_Category)","Value_Of_Care_Category")
facilityDf.show()
display(facilityDf)
```

```
# confirm changes
display(facilityDf)
```

```
# add Payment_ID for SQL DB
from pyspark.sql.window import Window as W
from pyspark.sql import functions as F
df = facilityDf.withColumn("Payment_ID", F.monotonically_increasing_id())

display(df)
```

```
# correct order of Payment_ID
from pyspark.sql.window import Window as W
from pyspark.sql import functions as F

windowSpec = W.orderBy("Facility_ID")
facilityDf = df.withColumn("Payment_ID", F.row_number().over(windowSpec))
facilityDf.show()
```
