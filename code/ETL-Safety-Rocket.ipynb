```
#read in file
df = spark.read.options(header='True').csv('/mnt/alysia/Safetyetl/safety.csv')
display(df)
```

```
# print column names
print(df.columns)
```

```
# dropping unnecessary columns as documented in Exercise-Initial-Capstone-Data-Sources
# viewing dataframe to make sure correct columns are dropped

df = df.select('facility_id', 'state', 'zip_code', 'hai6_baseline_rate', 'hai6_performance_rate')
df.show()
display(df)
```

```
# renaming columns for SQL database/readaibility
# confirming dataframe accepted column name changes

df = df.withColumnRenamed("facility_id","Facility_ID").withColumnRenamed("zip_code","Zip_Code").withColumnRenamed("hai6_baseline_rate","HAI_Baseline").withColumnRenamed("hai6_performance_rate","HAI_Performance")
df.show()
display(df)
```

```
# view current data types
df.printSchema()
```

```
# change data types for appropriate columns
# int: zip code
# double: all baseline and performance score columns
# display dataframe and print schema to confirm changes accepted
df = df.withColumn('Zip_Code', df.Zip_Code.cast('int')).withColumn('HAI_Baseline', df.HAI_Baseline.cast('double')).withColumn('HAI_Performance', df.HAI_Performance.cast('double'))
df.show()
display(df)
df.printSchema()
```

```
# dealing with missing values/nulls by imputing the median value of the respective column
from pyspark.sql.functions import col

# filter numeric cols
num_cols = [col_type[0] for col_type in filter(lambda dtype: dtype[1] in {"bigint", "double", "int"}, df.dtypes)]
# compute a dict with <col_name, median_value>
median_dict = dict()
for c in num_cols:
   median_dict[c] = df.stat.approxQuantile(c, [0.5], 0.001)[0]
df_imputed = df.na.fill(median_dict)
# check to make sure changes were made
# view finalized data frame
display(df_imputed)
```

```
# drop columns for SQL 
df_imputed1 = df_imputed.drop('State', 'Zip_Code')
df_imputed1.show()
display(df_imputed1)
```

```
# add Safety_ID for SQL DB
from pyspark.sql.window import Window as W
from pyspark.sql import functions as F
df = df_imputed1.withColumn("Safety_ID", F.monotonically_increasing_id())

display(df)
```

```
# correct order of Safety_ID
from pyspark.sql.window import Window as W
from pyspark.sql import functions as F

windowSpec = W.orderBy("Facility_ID")
df_imputed1 = df.withColumn("Safety_ID", F.row_number().over(windowSpec))
df_imputed1.show()
```
